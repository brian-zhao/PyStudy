### Key
The object name, e.g. xxx.jpg

### Value
This is the data itself, which is made up of asequence of bytes.

### Version ID
Allow you store multiple versions of the same object.

### Metadata
Data about the data you are storing, e.g. content-type, last-modified, etc.


### Securing S3

* Secure by default
* Control access to buckets (Bucket policies, Access control lists)
* Access logs (no configured by default, enable by yourself)




### S3 Encryption

* Encryption in Transit
    * SSL/TSL
    * HTTPS
* Encryption At Rest - Server Side Encryption
    * SSE-S3 (S3 managed keys, using AES 266 bit encryption)
    * SSE-KMS (AWS key management service managed keys)
    * SSE-C (Customer provided keys)
* Client Side Encryption
    * You encrypt the files yourself before you upload them into S3.

SSE:
    * Concole
    * Bucket Policy
        * Put request
        * x-amz-server-side-encryption: AES256
        * x-amz-server-side-encryption: aws:kms

## SQS

* Buffer between the component receiving and producing
* Pull Based
* 1min to 14 days

## SQS Queue Types
* Standard Queue best-effort ordering
* FIFO queue order preserved

Standard
* delivered at least once
* duplicates

FIFO
* No duplicates
* Deliver once

## SQS Settings
* Visibility Timeout
* amount of time that the message is invisible in the SQS queue.
* default timeout 30 second
* can up to 12 hours.
* Short polling (returns a response immediately even if the message queue being polled is empty)
* Long Polling (Periodically polls the queue, do not return a response until message arrives in the message queue or the long poll times out, 20 seconds maximum)

## SQS Delay Queues & Large Messages
Delay queues
* postpone delivery of new message
* default 0 sec to 900 sec
* FIFO queue delay the message

Large message
* use S3 (from 256KB to 2GB)
* SQS extended client library for Java
* AWS SDK for java
